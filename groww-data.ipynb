```python
df['date'] = pd.to_datetime(df['date'])
df['time'] = df['date'].dt.time

# Strategy parameters
ENTRY_TIME = time(9, 20)
EXIT_TIME = time(15, 10)
STOP_LOSS_PCT = 0.20  # 20% stop loss

# Initialize results storage
trades = []

# Get unique trading days
df['trading_day'] = df['date'].dt.date
trading_days = df['trading_day'].unique()

for day in trading_days:
    day_data = df[df['trading_day'] == day].copy()
    day_data = day_data.sort_values('date')
    
    # Find entry point (9:20)
    entry_rows = day_data[day_data['time'] == ENTRY_TIME]
    if entry_rows.empty:
        continue
    
    entry_idx = entry_rows.index[0]
    entry_price = day_data.loc[entry_idx, 'close']
    
    # Assuming ATM strike is closest to entry price (rounded to nearest 50)
    atm_strike = round(entry_price / 50) * 50
    
    # Simulate option premiums (you'll need actual option data)
    # For demonstration, assuming CE and PE premiums are 2% of spot each
    ce_entry_premium = atm_strike * 0.02
    pe_entry_premium = atm_strike * 0.02
    
    # Calculate stop loss prices
    ce_sl_price = ce_entry_premium * (1 + STOP_LOSS_PCT)
    pe_sl_price = pe_entry_premium * (1 + STOP_LOSS_PCT)
    
    # Track if stop loss hit
    ce_sl_hit = False
    pe_sl_hit = False
    ce_exit_premium = ce_entry_premium
    pe_exit_premium = pe_entry_premium
    exit_reason = "Time Exit"
    
    # Get data after entry
    after_entry = day_data[day_data.index > entry_idx]
    
    for idx, row in after_entry.iterrows():
        current_time = row['time']
        spot_price = row['close']
        
        # Simulate current option premiums based on spot movement
        # Simple approximation: premium changes with delta ~0.5
        spot_change_pct = (spot_price - entry_price) / entry_price
        
        # CE premium increases when spot goes up
        ce_current = ce_entry_premium * (1 + spot_change_pct * 0.5)
        # PE premium increases when spot goes down
        pe_current = pe_entry_premium * (1 - spot_change_pct * 0.5)
        
        # Add some volatility (random component)
        ce_current = max(ce_current, ce_entry_premium * 0.5)
        pe_current = max(pe_current, pe_entry_premium * 0.5)
        
        # Check CE stop loss
        if not ce_sl_hit and ce_current >= ce_sl_price:
            ce_sl_hit = True
            ce_exit_premium = ce_current
            exit_reason = "CE SL Hit"
        
        # Check PE stop loss
        if not pe_sl_hit and pe_current >= pe_sl_price:
            pe_sl_hit = True
            pe_exit_premium = pe_current
            if exit_reason == "CE SL Hit":
                exit_reason = "Both SL Hit"
            else:
                exit_reason = "PE SL Hit"
        
        # Exit at 15:10 or if both SL hit
        if current_time >= EXIT_TIME:
            if not ce_sl_hit:
                ce_exit_premium = ce_current
            if not pe_sl_hit:
                pe_exit_premium = pe_current
            break
    
    # Calculate P&L (selling options, so profit when premium decreases)
    ce_pnl = ce_entry_premium - ce_exit_premium
    pe_pnl = pe_entry_premium - pe_exit_premium
    total_pnl = ce_pnl + pe_pnl
    
    # Record trade
    trades.append({
        'Date': day,
        'Entry_Time': ENTRY_TIME,
        'Entry_Spot': entry_price,
        'ATM_Strike': atm_strike,
        'CE_Entry_Premium': ce_entry_premium,
        'PE_Entry_Premium': pe_entry_premium,
        'CE_Exit_Premium': ce_exit_premium,
        'PE_Exit_Premium': pe_exit_premium,
        'CE_PnL': ce_pnl,
        'PE_PnL': pe_pnl,
        'Total_PnL': total_pnl,
        'Exit_Reason': exit_reason
    })

# Create results DataFrame
results_df = pd.DataFrame(trades)

# Calculate statistics
if not results_df.empty:
    print("=" * 80)
    print("BACKTEST RESULTS - ATM STRADDLE SELLING STRATEGY")
    print("=" * 80)
    print(f"\nStrategy: Sell 1 ATM CE + 1 ATM PE at 9:20, Exit at 15:10")
    print(f"Stop Loss: {STOP_LOSS_PCT * 100}% on each leg")
    print(f"\nTotal Trading Days: {len(results_df)}")
    print(f"\nTotal P&L: ₹{results_df['Total_PnL'].sum():.2f}")
    print(f"Average P&L per Trade: ₹{results_df['Total_PnL'].mean():.2f}")
    print(f"Win Rate: {(results_df['Total_PnL'] > 0).sum() / len(results_df) * 100:.2f}%")
    print(f"Max Profit: ₹{results_df['Total_PnL'].max():.2f}")
    print(f"Max Loss: ₹{results_df['Total_PnL'].min():.2f}")
    print(f"\nExit Reasons:")
    print(results_df['Exit_Reason'].value_counts())
    print("\n" + "=" * 80)
    
    # Display trade log
    print("\nTRADE LOG:")
    print(results_df.to_string(index=False))
    
    # Save results
    #results_df.to_csv('backtest_results.csv', index=False)
    print("\nResults saved to 'backtest_results.csv'")
else:
    print("No trades executed. Check your data and entry time.")
```

```python
import pandas as pd
from datetime import datetime, timedelta
import time

def get_nifty_historical_options_data(groww, years=[2025]):
    """
    Get historical option data for NIFTY options across multiple years.
    
    Parameters:
    -----------
    groww : object
        Groww API client instance
    years : list
        List of years to fetch data for (default: [2024, 2025])
    
    Returns:
    --------
    pd.DataFrame
        Combined dataframe with all historical option data
    """
    
    all_data = []
    
    for year in years:
        print(f"Processing year: {year}")
        
        try:
            # Get all expiry dates for the year
            nifty_expiries = groww.get_expiries(
                exchange=groww.EXCHANGE_NSE,
                underlying_symbol="NIFTY",
                year=year,
                month=1
            )
            
            print(f"Found {len(nifty_expiries)} expiries for {year}")
            
            for expiry_date in nifty_expiries['expiries']:
                print(f"  Processing expiry: {expiry_date}")
                
                try:
                    # Get all contracts for this expiry
                    nifty_contracts = groww.get_contracts(
                        exchange=groww.EXCHANGE_NSE,
                        underlying_symbol="NIFTY",
                        expiry_date=expiry_date
                    )
                    
                    print(f"    Found {len(nifty_contracts)} contracts")
                    
                    # Parse expiry date
                    expiry_dt = datetime.strptime(expiry_date, '%Y-%m-%d')
                    
                    # Set start time (you can adjust this based on your needs)
                    # For example: start from 1 week before expiry
                    start_dt = expiry_dt - timedelta(days=7)
                    end_dt = expiry_dt
                    
                    # Format dates for API
                    start_time = start_dt.strftime('%Y-%m-%d 09:15:00')
                    end_time = end_dt.strftime('%Y-%m-%d 15:30:00')
                    
                    for contract in nifty_contracts:
                        # Extract contract details
                        groww_symbol = contract.get('groww_symbol') or contract.get('symbol')
                        strike_price = contract.get('strike_price')
                        option_type = contract.get('option_type')
                        
                        print(f"      Fetching data for: {groww_symbol}")
                        
                        try:
                            # Get historical candle data
                            historical_candles_response = groww.get_historical_candles(
                                exchange=groww.EXCHANGE_NSE,
                                segment=groww.SEGMENT_FNO,
                                groww_symbol=groww_symbol,
                                start_time=start_time,
                                end_time=end_time,
                                candle_interval=groww.CANDLE_INTERVAL_MIN_5
                            )
                            
                            # Convert to dataframe
                            if historical_candles_response and len(historical_candles_response) > 0:
                                df = pd.DataFrame(historical_candles_response)
                                
                                # Add contract metadata
                                df['groww_symbol'] = groww_symbol
                                df['strike_price'] = strike_price
                                df['option_type'] = option_type
                                df['expiry_date'] = expiry_date
                                df['year'] = year
                                
                                all_data.append(df)
                                print(f"        Added {len(df)} candles")
                            else:
                                print(f"        No data available")
                            
                            # Add a small delay to avoid rate limiting
                            time.sleep(0.5)
                            
                        except Exception as e:
                            print(f"        Error fetching candles for {groww_symbol}: {str(e)}")
                            continue
                    
                except Exception as e:
                    print(f"    Error fetching contracts for {expiry_date}: {str(e)}")
                    continue
                    
        except Exception as e:
            print(f"Error fetching expiries for {year}: {str(e)}")
            continue
    
    # Combine all dataframes
    if all_data:
        final_df = pd.concat(all_data, ignore_index=True)
        
        # Convert timestamp column if it exists
        if 'timestamp' in final_df.columns:
            final_df['timestamp'] = pd.to_datetime(final_df['timestamp'])
        elif 'date' in final_df.columns:
            final_df['date'] = pd.to_datetime(final_df['date'])
        
        # Sort by date and symbol
        sort_cols = ['timestamp'] if 'timestamp' in final_df.columns else ['date']
        sort_cols.extend(['groww_symbol'])
        final_df = final_df.sort_values(sort_cols).reset_index(drop=True)
        
        print(f"\n✓ Total records: {len(final_df)}")
        print(f"✓ Unique contracts: {final_df['groww_symbol'].nunique()}")
        print(f"✓ Date range: {final_df[sort_cols[0]].min()} to {final_df[sort_cols[0]].max()}")
        
        return final_df
    else:
        print("No data collected")
        return pd.DataFrame()


# Usage example:
# df_options = get_nifty_historical_options_data(groww, years=[2024, 2025])

# Optional: Save to CSV
# df_options.to_csv('nifty_options_historical_data.csv', index=False)

# Optional: Filter specific option types
# df_ce = df_options[df_options['option_type'] == 'CE']  # Call options
# df_pe = df_options[df_options['option_type'] == 'PE']  # Put options
```

```python
def get_nifty_historical_options_data(groww, years=[2025]):
    """
    Get historical option data for NIFTY options with rate limit handling
    """
    all_data = []
    
    # Rate limiting parameters
    REQUEST_DELAY = 1  # Base delay between requests in seconds
    MAX_RETRIES = 3    # Maximum number of retries per request
    BACKOFF_FACTOR = 2 # Multiply delay by this factor after each failure
    
    def make_api_request_with_retry(func, *args, **kwargs):
        """Helper function to make API requests with retry logic"""
        for attempt in range(MAX_RETRIES):
            try:
                response = func(*args, **kwargs)
                time.sleep(REQUEST_DELAY)  # Basic rate limiting
                return response
            except Exception as e:
                if attempt < MAX_RETRIES - 1:
                    wait_time = REQUEST_DELAY * (BACKOFF_FACTOR ** attempt)
                    print(f"Request failed, retrying in {wait_time} seconds...")
                    time.sleep(wait_time)
                else:
                    raise e
    
    for year in years:
        print(f"Processing year: {year}")
        
        # Get expiries with retry logic
        nifty_expiries = make_api_request_with_retry(
            groww.get_expiries,
            exchange=groww.EXCHANGE_NSE,
            underlying_symbol="NIFTY",
            year=year,
            month=2
        )
        
        if not nifty_expiries or 'expiries' not in nifty_expiries:
            print(f"No expiries found for {year}")
            continue
            
        print(f"Found {len(nifty_expiries['expiries'])} expiries for {year}")
        
        for expiry_date in nifty_expiries['expiries']:
            print(f"Processing expiry: {expiry_date}")
                
            # Get contracts with retry logic
            nifty_contracts = make_api_request_with_retry(
                groww.get_contracts,
                exchange=groww.EXCHANGE_NSE,
                underlying_symbol="NIFTY",
                expiry_date=expiry_date
            )
            
            if not nifty_contracts or 'contracts' not in nifty_contracts:
                print(f"No contracts found for {expiry_date}")
                continue
                
            print(f"Found {len(nifty_contracts['contracts'])} contracts")
            
            # Calculate date range
            expiry_dt = datetime.strptime(expiry_date, '%Y-%m-%d')
            start_dt = expiry_dt - timedelta(days=7)
            end_dt = expiry_dt
            
            start_time = start_dt.strftime('%Y-%m-%d 09:15:00')
            end_time = end_dt.strftime('%Y-%m-%d 15:30:00')
            
            # Process only a subset of contracts if needed
            # contracts_to_process = nifty_contracts['contracts'][:10]  # Uncomment to limit contracts
            
            for contract in nifty_contracts['contracts']:
                print(f"Fetching data for: {contract}")
    return nifty_contracts
```

```python
def get_nifty_historical_options_data(groww, years=[2025]):
    """
    Get historical option data for NIFTY options with rate limit handling
    """
    all_data = []
    
    # Rate limiting parameters
    REQUEST_DELAY = 1  # Base delay between requests in seconds
    MAX_RETRIES = 3    # Maximum number of retries per request
    BACKOFF_FACTOR = 2 # Multiply delay by this factor after each failure
    
    def make_api_request_with_retry(func, *args, **kwargs):
        """Helper function to make API requests with retry logic"""
        for attempt in range(MAX_RETRIES):
            try:
                response = func(*args, **kwargs)
                time.sleep(REQUEST_DELAY)  # Basic rate limiting
                return response
            except Exception as e:
                if attempt < MAX_RETRIES - 1:
                    wait_time = REQUEST_DELAY * (BACKOFF_FACTOR ** attempt)
                    print(f"Request failed, retrying in {wait_time} seconds...")
                    time.sleep(wait_time)
                else:
                    raise e
    
    for year in years:
        print(f"Processing year: {year}")
        
        try:
            # Get expiries with retry logic
            nifty_expiries = make_api_request_with_retry(
                groww.get_expiries,
                exchange=groww.EXCHANGE_NSE,
                underlying_symbol="NIFTY",
                year=year,
                month=9
            )
            
            if not nifty_expiries or 'expiries' not in nifty_expiries:
                print(f"No expiries found for {year}")
                continue
                
            print(f"Found {len(nifty_expiries['expiries'])} expiries for {year}")
            
            for expiry_date in nifty_expiries['expiries']:
                print(f"Processing expiry: {expiry_date}")
                
                try:
                    # Get contracts with retry logic
                    nifty_contracts = make_api_request_with_retry(
                        groww.get_contracts,
                        exchange=groww.EXCHANGE_NSE,
                        underlying_symbol="NIFTY",
                        expiry_date=expiry_date
                    )
                    
                    if not nifty_contracts or 'contracts' not in nifty_contracts:
                        print(f"No contracts found for {expiry_date}")
                        continue
                        
                    print(f"Found {len(nifty_contracts['contracts'])} contracts")
                    
                    # Calculate date range
                    expiry_dt = datetime.strptime(expiry_date, '%Y-%m-%d')
                    start_dt = expiry_dt - timedelta(days=7)
                    end_dt = expiry_dt
                    
                    start_time = start_dt.strftime('%Y-%m-%d 09:15:00')
                    end_time = end_dt.strftime('%Y-%m-%d 15:30:00')
                    
                    # Process only a subset of contracts if needed
                    # contracts_to_process = nifty_contracts['contracts'][:10]  # Uncomment to limit contracts
                    
                    for contract in nifty_contracts['contracts']:
                        print(f"Fetching data for: {contract}")
                        
                        try:
                            # Get historical data with retry logic
                            candles = make_api_request_with_retry(
                                groww.get_historical_candles,
                                exchange=groww.EXCHANGE_NSE,
                                segment=groww.SEGMENT_FNO,
                                groww_symbol=contract,
                                start_time=start_time,
                                end_time=end_time,
                                candle_interval=groww.CANDLE_INTERVAL_MIN_5
                            )
                            
                            if candles and 'candles' in candles and len(candles['candles']) > 0:
                                df = pd.DataFrame(candles['candles'], columns=[
                                    'timestamp', 'open', 'high', 'low', 'close', 
                                    'volume', 'oi'
                                ])
                                
                                # Add metadata
                                df['groww_symbol'] = contract
                                df['strike_price'] = int(contract.split('-')[-2])
                                df['option_type'] = contract.split('-')[-1]
                                df['expiry'] = contract.split('-')[-3]
                                df['expiry_date'] = expiry_date
                                
                                all_data.append(df)
                                print(f"Added {len(df)} candles")
                            
                        except Exception as e:
                            print(f"Error processing contract {contract}: {str(e)}")
                            continue
                            
                except Exception as e:
                    print(f"Error processing expiry {expiry_date}: {str(e)}")
                    continue
                    
        except Exception as e:
            print(f"Error processing year {year}: {str(e)}")
            continue
    
    if not all_data:
        print("No data collected")
        return pd.DataFrame()
    
    # Combine and process final dataframe
    final_df = pd.concat(all_data, ignore_index=True)
    
    # Convert timestamp and add calculated fields
    final_df['timestamp'] = pd.to_datetime(final_df['timestamp'])
    final_df['date'] = pd.to_datetime(final_df['timestamp'])
    final_df['expiry_date'] = pd.to_datetime(final_df['expiry_date'])
    
    # Calculate DTE (Days to Expiry)
    final_df['dte_num'] = (final_df['expiry_date'] - final_df['date']).dt.days + 1
    final_df['DTE'] = final_df['dte_num'].apply(
        lambda x: 'ODTE' if x == 0 else f'{x}DTE' if x <= 7 else '>7DTE'
    )
    
    # Clean up and sort
    final_df = final_df.drop('dte_num', axis=1)
    final_df = final_df.sort_values(['timestamp', 'groww_symbol']).reset_index(drop=True)
    
    print(f"\nProcessing complete:")
    print(f"Total records: {len(final_df)}")
    print(f"Unique contracts: {final_df['groww_symbol'].nunique()}")
    print(f"Date range: {final_df['timestamp'].min()} to {final_df['timestamp'].max()}")
    
    return final_df

# Usage:
df = get_nifty_historical_options_data(groww, years=[2024])
```
